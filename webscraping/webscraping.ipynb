```python
#https://www.youtube.com/watch?v=QdLgNr1mKQU&t=26s

import csv
import requests
from bs4 import BeautifulSoup
```

```python
# URL da página que queremos fazer scraping
url = "https://www.mercadolivre.com.br/ofertas#nav-header"
```

```python
# Conexão: Enviar uma solicitação GET para a URL
response = requests.get(url)
```

```python
if response.status_code == 200:
    # Parse a página com o BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Encontre os elementos HTML que contêm os títulos de notícias
    products = soup.find_all("div", class_="promotion-item__description")
    print(products)
    
    #cria arquivo csv
    file = open('export_data.csv', 'w', newline='')
    writer = csv.writer(file)
    headers = ['Produto', 'Preço']
    writer.writerow(headers)

    # Loop pelos elementos e imprimir os títulos
    for product in products:

        title = product.find('p', class_='promotion-item__title')
        price = product.find('span', class_='andes-money-amount__fraction')
       
        print(title.text, price.text)
        #cada noticia
        item = [title.text, price.text]

        #salvar noticia no arquivo
        file = open('export_data.csv', 'a', newline='', encoding='utf-8')
        writer = csv.writer(file)
        writer.writerow(item)
        file.close()

else:
    print("Falha ao acessar a página:", response.status_code)
```
